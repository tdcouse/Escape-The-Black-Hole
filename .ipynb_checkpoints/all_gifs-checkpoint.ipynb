# Captured by Aliens

```python
#Escape the Black Hole game -- Aliens catch you
import numpy as np
import matplotlib.pyplot as py
from scipy.integrate import odeint


def deriv(s, t, G, mBH, mA, mS, mSD):
  xS0 = s[0]
  vxS0 = s[1]
  yS0 = s[2]
  vyS0 = s[3]
  xBH0 = s[4]
  vxBH0 = s[5]
  yBH0 = s[6]
  vyBH0 = s[7]
  xA0 = s[8]
  vxA0 = s[9]
  yA0 = s[10]
  vyA0 = s[11]
  xSD0 = s[12]
  vxSD0 = s[13]
  ySD0 = s[14]
  vySD0 = s[15]

  f1 = vxS0
  f2 = ((G*mA*(xA0-xS0))/(((xA0-xS0)**2+(yA0-yS0)**2)**(1.5))) + ((G*mBH*(xBH0-xS0))/(((xBH0-xS0)**2+(yBH0-yS0)**2)**(1.5)))
  f3 = vyS0
  f4 = ((G*mA*(yA0-yS0))/(((xA0-xS0)**2+(yA0-yS0)**2)**(1.5))) + ((G*mBH*(yBH0-yS0))/(((xBH0-xS0)**2+(yBH0-yS0)**2)**(1.5)))
  f5 = vxBH0
  f6 = ((G*mA*(xA0-xBH0))/(((xA0-xBH0)**2+(yA0-yBH0)**2)**(1.5))) + ((G*mS*(xS0-xBH0))/(((xS0-xBH0)**2+(yS0-yBH0)**2)**(1.5)))
  f7 = vyBH0
  f8 = ((G*mA*(yA0-yBH0))/(((xA0-xBH0)**2+(yA0-yBH0)**2)**(1.5))) + ((G*mS*(yS0-yBH0))/(((xS0-xBH0)**2+(yS0-yBH0)**2)**(1.5)))
  f9 = vxA0
  f10 = ((G*mS*(xS0-xA0))/(((xS0-xA0)**2+(yS0-yA0)**2)**(1.5))) + ((G*mBH*(xBH0-xA0))/(((xBH0-xA0)**2+(yBH0-yA0)**2)**(1.5)))
  f11 = vyA0
  f12 = ((G*mS*(yS0-yA0))/(((xS0-xA0)**2+(yS0-yA0)**2)**(1.5))) + ((G*mBH*(yBH0-yA0))/(((xBH0-xA0)**2+(yBH0-yA0)**2)**(1.5)))
  f13 = vxSD0
  f14 = ((G*mSD*(xA0-xSD0))/(((xA0-xSD0)**2+(yA0-ySD0)**2)**(1.5))) + ((G*mBH*(xBH0-xSD0))/(((xBH0-xSD0)**2+(yBH0-ySD0)**2)**(1.5)))
  f15 = vySD0
  f16 = ((G*mSD*(yA0-ySD0))/(((xA0-xSD0)**2+(yA0-ySD0)**2)**(1.5))) + ((G*mBH*(yBH0-ySD0))/(((xBH0-xSD0)**2+(yBH0-ySD0)**2)**(1.5)))
  
  #We are saying the alien ship is no longer a death star, but instead a star destroyer, a second large alien space ship
  
  
  return [f1, f2, f3, f4, f5, f6, f7, f8, f9, f10, f11, f12, f13, f14, f15, f16]

def diffeq_solver_from_scipy(s_initial, tmin, tmax, nts, deriv, params):
    t = np.linspace(tmin,tmax,nts)
    s = odeint(deriv, s_initial, t, params)
    
    return t,s

G = 6.67430*(10**(-11))
mBH = 5.1714*(10**36)#Mass Sagitatius A
mA = 2.5*(10**18)#Mass of the Death Star
mS = 3.0*(10**6)#Mass of Saturn V
mSD = mA = 100.0*(10**6)#Mass of star destroyer

#Schwarzschild radius of Sagitarius A is 7.86e9 m

xS0 = 0.0
vxS0 = (0.21*(3*(10**8))) #m/s #this works at .01c but we want a bigger curve so we changed it
yS0 = 0.786*(10**(10))
#vyS0 = (0.96)*(3*(10**8)) #m/s - choose this to escape
vyS0 = (0.95)*(3*(10**8)) #m/s up to this speed will never escape the black hole
#vyS0 = (0.988)*(3*(10**8)) #m/s minimum speed to escape, there is not speed possible above 1.0*c - for most calculating plot above, not this particlar small scale view
xBH0 = 0.0
vxBH0 = 0.0
yBH0 = 0.0
vyBH0 = 0.0
xA0 = -8.7*(10**(10))
vxA0 = (0.01)*(3*(10**8))
yA0 = 0.0
vyA0 =  np.sqrt(G*mBH/np.abs(xA0))
xSD0 = 16.7*(10**(10))
vxSD0 = -(0.085)*(3*(10**8))
ySD0 = 2.7*(10**(10))
vySD0 =  -(0.117*(3*(10**8)))

s_initial = [xS0, vxS0, yS0, vyS0,xBH0, vxBH0, yBH0, vyBH0, xA0, vxA0, yA0, vyA0, xSD0, vxSD0, ySD0, vySD0]
tmin = 0
tmax = 10319
nts = 250

params = (G, mBH, mA, mS, mSD)

(t,s) = diffeq_solver_from_scipy(s_initial, tmin, tmax, nts, deriv, params)


xS = s[:,0]
vxS = s[:,1]
yS = s[:,2]
vyS = s[:,3]
xBH = s[:,4]
vxBH = s[:,5]
yBH = s[:,6]
vyBH = s[:,7]
xA = s[:,8]
vxA = s[:,9]
yA = s[:,10]
vyA = s[:,11]
xSD = s[:,12]
vxSD = s[:,13]
ySD = s[:,14]
vySD = s[:,15]

circle1 = py.Circle((0, 0), 7.8*(10**(9)), color='k')
img = py.imread("crazy2.jpg")

py.plot(xS, yS, 'r3',label='Ship Orbit')
py.plot(xA, yA, 'm.', label='UFO')
py.plot(xBH, yBH,'kX',label='Singularity')
py.plot(xSD, ySD,'c4',label='Hostile Aliens')
py.legend(loc='lower right')
fig = py.gcf()
ax = fig.gca()
py.title("Expedition Escape the Black Hole")
py.xlabel("X position (m)")
py.ylabel("Y position (m)")
py.xlim(-15.5*(10**(10)), 15.5*(10**(10)))
py.ylim(-20.5*(10**(10)), 25.5*(10**(10)))
ax.add_artist(circle1)
ax.imshow(img,extent=[-27.5*(10**(10)), 31.5*(10**(10)), -29.5*(10**(10)), 25.5*(10**(10))])
py.rcParams["figure.figsize"] = (16, 15)
py.show()
```
